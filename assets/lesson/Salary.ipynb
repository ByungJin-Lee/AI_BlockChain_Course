```python
#salary

import pandas as pd
import numpy as np

s_df = pd.read_csv('salary.csv')
s_df.head(10)
```

```python
s_df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

def labelencoding(df):
    for c in df.columns:
        if df[c].dtype == 'object':
            df[c] = df[c].fillna('N')
            lbl = LabelEncoder()
            lbl.fit(list(df[c].values))
            df[c] = lbl.transform(df[c].values)
    return df
```

```python
l_df = labelencoding(s_df)
l_df.head(10)

s_labels = l_df['salary']
s_datas = l_df.drop('salary', axis=1)
```

## NAN

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(s_datas)
s_datas = imputer.transform(s_datas)
```

## Scaler

```python
from sklearn.preprocessing import MinMaxScaler
scl = MinMaxScaler(feature_range=(0,1))
s_datas = scl.fit_transform(s_datas)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(s_datas, s_labels, test_size=0.2, random_state=12)
```

```python
dt = RandomForestClassifier(random_state=12)
dt.fit(x_train, y_train)
```

```python
pred = dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print(classification_report(y_test, pred))
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test, pred))
```
