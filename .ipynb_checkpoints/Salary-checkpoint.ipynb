```python
#salary

import pandas as pd

s_df = pd.read_csv('salary.csv')
s_df.head(10)
```

```python
s_df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

def labelencoding(df):
    for c in df.columns:
        if df[c].dtype == 'object':
            df[c] = df[c].fillna('N')
            lbl = LabelEncoder()
            lbl.fit(list(df[c].values))
            df[c] = lbl.transform(df[c].values)
    return df
```

```python
l_df = labelencoding(s_df)
l_df.head(10)
l_df.fillna(0)
```

```python
s_labels = l_df['salary'].values
s_datas = l_df.drop('salary', axis=1).values

print(s_labels, s_datas)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(s_datas, s_labels, test_size=0.2, random_state=12)
```

```python
dt = DecisionTreeClassifier(random_state=12)
dt.fit(x_train, y_train)
```
