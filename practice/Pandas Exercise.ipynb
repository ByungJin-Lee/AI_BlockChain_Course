### Exercise for Pandas

```python
#import Pandas, numpy module
import pandas as pd
import numpy as np
```

```python
#make a dataframe from dict
p1_dict = {
    '이름': ['홍길동', '에밀리', '이치로'],
    '출생지': ['한국', '미국', '일본'],
    '나이' : [20, 23, 25]
}

p1_df = pd.DataFrame(p1_dict)
p1_df
```

```python
#add column '성별'
p2_sex = ['남', '여', '남']
p1_df['성별'] = p2_sex
p2_df = p1_df
p2_df
```

```python
#add row '슈에리, 중국, 22, 여'
p3_row_dict = {
    '이름': '슈에리',
    '출생지': '중국',
    '나이': 22,
    '성별': '여'
}
p3_df = p2_df.append(p3_row_dict, ignore_index=True)
p3_df
```

```python
#get mean that student's age
p3_df['나이'].mean()
```

### Titanic

```python
#read csv and transform dataframe
titanic_df = pd.read_csv('titanic_train.csv')
titanic_df
```

```python
#get number of survived group by 'Pclass'
p6_df = titanic_df[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).count()
p6_df
```

```python
#get survived %
p7_df = titanic_df[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).mean()
p7_df
```

```python
p8_df = titanic_df['Age'].agg([max, min])
p8_df['max'], p8_df['min'], titanic_df['Age'].mean()
```
